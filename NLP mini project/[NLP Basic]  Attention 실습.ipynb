##### getBatch í•¨ìˆ˜ êµ¬í˜„

`getBatch` í•¨ìˆ˜ëŠ” ì£¼ì–´ì§„ `train_data`ì—ì„œ ë¬´ì‘ìœ„ë¡œ ì„ íƒëœ ë°°ì¹˜ (batch)ë¥¼ ìƒì„±í•˜ëŠ” ì—­í• ì„ í•©ë‹ˆë‹¤. ì´ í•¨ìˆ˜ëŠ” ë‹¤ìŒê³¼ ê°™ì€ ê³¼ì •ì„ ê±°ì³ ë°°ì¹˜ë¥¼ ìƒì„±í•©ë‹ˆë‹¤:

- `train_data`ë¥¼ ë¬´ì‘ìœ„ë¡œ ì„ìŠµë‹ˆë‹¤
- `sindex`ì™€ `eindex` ë³€ìˆ˜ë¥¼ ì‚¬ìš©í•˜ì—¬ `train_data`ì—ì„œ ë°°ì¹˜ë¥¼ ì¶”ì¶œí•©ë‹ˆë‹¤. `sindex`ëŠ” ì‹œì‘ ì¸ë±ìŠ¤ë¥¼ ë‚˜íƒ€ë‚´ê³ , `eindex`ëŠ” ë ì¸ë±ìŠ¤ë¥¼ ë‚˜íƒ€ëƒ…ë‹ˆë‹¤
- `eindex`ë¥¼ `batch_size`ë§Œí¼ ì¦ê°€ì‹œí‚¤ê³ , `sindex`ë¥¼ ì´ì „ `eindex` ê°’ìœ¼ë¡œ ì—…ë°ì´íŠ¸í•©ë‹ˆë‹¤
- `eindex`ê°€ `train_data`ì˜ ê¸¸ì´ë³´ë‹¤ ì‘ì€ ë™ì•ˆ ìœ„ ê³¼ì •ì„ ë°˜ë³µí•˜ë©° ë°°ì¹˜ë¥¼ ìƒì„±í•©ë‹ˆë‹¤
- `eindex`ê°€ `train_data`ì˜ ê¸¸ì´ë³´ë‹¤ í¬ê±°ë‚˜ ê°™ì•„ì§€ë©´, ë‚¨ì€ ë°ì´í„°ë¥¼ ë§ˆì§€ë§‰ ë°°ì¹˜ë¡œ ìƒì„±í•©ë‹ˆë‹¤

````python
# ë°°ì¹˜ ë°ì´í„°ë¥¼ ê°€ì ¸ì˜¤ëŠ” í•¨ìˆ˜
def getBatch(batch_size, train_data):
    random.shuffle(train_data) # ë°ì´í„°ë¥¼ ë¬´ì‘ìœ„ë¡œ ì„ìŒ
    sindex = 0 # ì‹œì‘ ì¸ë±ìŠ¤
    eindex = batch_size # ë ì¸ë±ìŠ¤
    while eindex < len(train_data): # ë ì¸ë±ìŠ¤ê°€ ë°ì´í„°ì˜ ê¸¸ì´ë³´ë‹¤ ì‘ì„ ë™ì•ˆ
        batch = train_data[sindex: eindex] # ë°°ì¹˜ ë°ì´í„°ë¥¼ ê°€ì ¸ì˜´
        temp = eindex
        eindex = eindex + batch_size # ë ì¸ë±ìŠ¤ë¥¼ ë°°ì¹˜ í¬ê¸°ë§Œí¼ ì¦ê°€
        sindex = temp # ì‹œì‘ ì¸ë±ìŠ¤ë¥¼ ì´ì „ ë ì¸ë±ìŠ¤ë¡œ ì—…ë°ì´íŠ¸
        yield batch # ë°°ì¹˜ ë°ì´í„° ë°˜í™˜

    if eindex >= len(train_data): # ë ì¸ë±ìŠ¤ê°€ ë°ì´í„°ì˜ ê¸¸ì´ë³´ë‹¤ í¬ê±°ë‚˜ ê°™ìœ¼ë©´
        batch = train_data[sindex:] # ë‚¨ì€ ë°ì´í„°ë¥¼ ë°°ì¹˜ë¡œ ë§Œë“¦
        yield batch # ë°°ì¹˜ ë°ì´í„° ë°˜í™˜
````

##### pad_to_batch í•¨ìˆ˜ êµ¬í˜„

ì´ í•¨ìˆ˜ëŠ” ì£¼ì–´ì§„ ë°°ì¹˜ì— íŒ¨ë”©ì„ ì ìš©í•˜ì—¬ ì…ë ¥ ì‹œí€€ìŠ¤ì™€ íƒ€ê²Ÿ ì‹œí€€ìŠ¤ë¥¼ ë™ì¼í•œ ê¸¸ì´ë¡œ ë§Œë“­ë‹ˆë‹¤. ì´ë¥¼ í†µí•´ ë°°ì¹˜ ì²˜ë¦¬ë¥¼ ìˆ˜í–‰í•  ìˆ˜ ìˆìŠµë‹ˆë‹¤

í•¨ìˆ˜ëŠ” ë‹¤ìŒ ë‹¨ê³„ë¥¼ ìˆ˜í–‰í•©ë‹ˆë‹¤:

1. ë°°ì¹˜ë¥¼ ê¸¸ì´ì— ë”°ë¼ ì •ë ¬
2. ì…ë ¥ ì‹œí€€ìŠ¤ì™€ íƒ€ê²Ÿ ì‹œí€€ìŠ¤ì˜ ìµœëŒ€ ê¸¸ì´ë¥¼ ê³„ì‚°
3. ê° ì‹œí€€ìŠ¤ì— íŒ¨ë”©ì„ ì ìš©í•˜ì—¬ ë™ì¼í•œ ê¸¸ì´ë¡œ ë³€í™˜
4. ì…ë ¥ ë³€ìˆ˜ì™€ íƒ€ê²Ÿ ë³€ìˆ˜ë¥¼ ìƒì„±í•˜ê³ , ê° ì‹œí€€ìŠ¤ì˜ ì‹¤ì œ ê¸¸ì´ë¥¼ ê³„ì‚°
5. ì…ë ¥ ë³€ìˆ˜, íƒ€ê²Ÿ ë³€ìˆ˜, ì…ë ¥ ê¸¸ì´, íƒ€ê²Ÿ ê¸¸ì´ë¥¼ ë°˜í™˜

ì´ í•¨ìˆ˜ë¥¼ ì‚¬ìš©í•˜ë©´, ì…ë ¥ ì‹œí€€ìŠ¤ì™€ íƒ€ê²Ÿ ì‹œí€€ìŠ¤ë¥¼ ë™ì¼í•œ ê¸¸ì´ë¡œ ë§Œë“¤ì–´ ë°°ì¹˜ ì²˜ë¦¬ë¥¼ ìˆ˜í–‰í•  ìˆ˜ ìˆìŠµë‹ˆë‹¤

````python
# ì‹œí€€ìŠ¤ë¥¼ ë°°ì¹˜ë¡œ ë³€í™˜í•˜ëŠ” í•¨ìˆ˜
def pad_to_batch(batch, x_to_ix, y_to_ix):
    sorted_batch =  sorted(batch, key=lambda b:b[0].size(1), reverse=True) # ê¸¸ì´ì— ë”°ë¼ ë°°ì¹˜ë¥¼ ì •ë ¬
    x,y = list(zip(*sorted_batch)) # ì…ë ¥ê³¼ ì¶œë ¥ì„ ë¶„ë¦¬
    max_x = max([s.size(1) for s in x]) # ì…ë ¥ì˜ ìµœëŒ€ ê¸¸ì´
    max_y = max([s.size(1) for s in y]) # ì¶œë ¥ì˜ ìµœëŒ€ ê¸¸ì´
    x_p, y_p = [], [] # íŒ¨ë”©ì„ ì ìš©í•  ì…ë ¥ê³¼ ì¶œë ¥ ë¦¬ìŠ¤íŠ¸

    for i in range(len(batch)):
        if x[i].size(1) < max_x: # ì…ë ¥ì˜ ê¸¸ì´ê°€ ìµœëŒ€ ê¸¸ì´ë³´ë‹¤ ì‘ìœ¼ë©´
            # íŒ¨ë”© í† í°ì„ ì¶”ê°€í•˜ì—¬ ìµœëŒ€ ê¸¸ì´ì— ë§ì¶¤
            x_p.append(torch.cat([x[i], Variable(LongTensor([x_to_ix['<PAD>']] * (max_x - x[i].size(1)))).view(1, -1)], 1))
        else:
            x_p.append(x[i]) # ê·¸ë ‡ì§€ ì•Šìœ¼ë©´ ê·¸ëŒ€ë¡œ ì‚¬ìš©
        if y[i].size(1) < max_y: # ì¶œë ¥ì˜ ê¸¸ì´ê°€ ìµœëŒ€ ê¸¸ì´ë³´ë‹¤ ì‘ìœ¼ë©´
            # íŒ¨ë”© í† í°ì„ ì¶”ê°€í•˜ì—¬ ìµœëŒ€ ê¸¸ì´ì— ë§ì¶¤
            y_p.append(torch.cat([y[i], Variable(LongTensor([y_to_ix['<PAD>']] * (max_y - y[i].size(1)))).view(1, -1)], 1))
        else:
            y_p.append(y[i]) # ê·¸ë ‡ì§€ ì•Šìœ¼ë©´ ê·¸ëŒ€ë¡œ ì‚¬ìš©

    input_var = torch.cat(x_p) # ì…ë ¥ ë³€ìˆ˜
    target_var = torch.cat(y_p) # ëª©í‘œ ë³€ìˆ˜
    input_len = [list(map(lambda s: s ==0, t.data)).count(False) for t in input_var] # ì…ë ¥ì˜ ì‹¤ì œ ê¸¸ì´
    target_len = [list(map(lambda s: s ==0, t.data)).count(False) for t in target_var] # ì¶œë ¥ì˜ ì‹¤ì œ ê¸¸ì´

    return input_var, target_var, input_len, target_len
````

##### ë°ì´í„°ì…‹ load ë° preprocess ì½”ë“œ êµ¬í˜„

1. `unicode_to_ascii` í•¨ìˆ˜ëŠ” ìœ ë‹ˆì½”ë“œ ë¬¸ìì—´ì„ ASCII ë¬¸ìì—´ë¡œ ë³€í™˜í•©ë‹ˆë‹¤. ì´ í•¨ìˆ˜ëŠ” ì£¼ì–´ì§„ ìœ ë‹ˆì½”ë“œ ë¬¸ìì—´ì—ì„œ ê° ë¬¸ìë¥¼ ìˆœíšŒí•˜ë©°, í•´ë‹¹ ë¬¸ìê°€ ë¹„ìˆ˜ì¹˜í˜•(non-mark)ì¸ ê²½ìš°ì—ë§Œ ASCII ë¬¸ìì—´ì— ì¶”ê°€í•©ë‹ˆë‹¤. ì´ë¥¼ í†µí•´ ìœ ë‹ˆì½”ë“œ ë¬¸ìì—´ì´ ASCII ë¬¸ìì—´ë¡œ ë³€í™˜ë©ë‹ˆë‹¤

<br>

2. `normalize_string` í•¨ìˆ˜ëŠ” ë¬¸ìì—´ì„ ì†Œë¬¸ìë¡œ ë³€í™˜í•˜ê³ , ê³µë°±ì„ ì œê±°í•˜ë©°, ì˜ë¬¸ìì™€ êµ¬ë‘ì  ì´ì™¸ì˜ ë¬¸ìë¥¼ ì œê±°í•©ë‹ˆë‹¤. ì´ í•¨ìˆ˜ëŠ” ë‹¤ìŒê³¼ ê°™ì€ ìˆœì„œë¡œ ì‘ë™í•©ë‹ˆë‹¤:
* ë¬¸ìì—´ì„ ì†Œë¬¸ìë¡œ ë³€í™˜í•˜ê³  ì–‘ìª½ ê³µë°±ì„ ì œê±°
* êµ¬ë‘ì  ì•ë’¤ì— ê³µë°±ì„ ì¶”ê°€í•˜ì—¬ êµ¬ë‘ì ê³¼ ë‹¨ì–´ê°€ êµ¬ë¶„ë˜ë„ë¡ í•¨
* ì˜ë¬¸ì, êµ¬ë‘ì  ì´ì™¸ì˜ ë¬¸ìë¥¼ ì œê±°
* ì—°ì†ëœ ê³µë°±ì„ í•˜ë‚˜ì˜ ê³µë°±ìœ¼ë¡œ ì¹˜í™˜í•˜ê³ , ë¬¸ìì—´ì˜ ì•ë’¤ ê³µë°±ì„ ì œê±°

ì´ ë‘ í•¨ìˆ˜ë¥¼ ì‚¬ìš©í•˜ì—¬ ë¬¸ìì—´ì„ ì²˜ë¦¬í•˜ê³  ì •ê·œí™”í•  ìˆ˜ ìˆìŠµë‹ˆë‹¤. ì •ê·œí™”ëœ ë°ì´í„°ëŠ” ëª¨ë¸ì˜ ì…ë ¥ ë°ì´í„°ë¥¼ ì ì ˆí•œ í˜•íƒœë¡œ ì‚¬ìš©ë©ë‹ˆë‹¤

````python
# ìœ ë‹ˆì½”ë“œ ë¬¸ìì—´ì„ ì•„ìŠ¤í‚¤ ë¬¸ìì—´ë¡œ ë³€í™˜í•˜ëŠ” í•¨ìˆ˜
def unicode_to_ascii(s):
    return ''.join(
        c for c in unicodedata.normalize('NFD', s)
        if unicodedata.category(c) != 'Mn'
    )

# ë¬¸ìì—´ì„ ì •ê·œí™”í•˜ëŠ” í•¨ìˆ˜
def normalize_string(s):
    s = unicode_to_ascii(s.lower().strip())
    s = re.sub(r"([,.!?])", r" \1 ", s)
    s = re.sub(r"[^a-zA-Z,.!?]+", r" ", s)
    s = re.sub(r"\s+", r" ", s).strip()
    return s
````

ì´ ì½”ë“œì—ì„œëŠ” ê¸°ê³„ ë²ˆì—­ì„ ìœ„í•œ ë°ì´í„° ì „ì²˜ë¦¬ì™€ í•„í„°ë§ì„ ìˆ˜í–‰í•©ë‹ˆë‹¤

ë¨¼ì €, ìµœì†Œ ë° ìµœëŒ€ ê¸¸ì´ë¥¼ ì„¤ì •í•©ë‹ˆë‹¤. ì´ë¥¼ í†µí•´ ë„ˆë¬´ ì§§ê±°ë‚˜ ë„ˆë¬´ ê¸´ ë¬¸ì¥ì„ ì œì™¸í•  ìˆ˜ ìˆìŠµë‹ˆë‹¤

- `MIN_LENGTH` : ë¬¸ì¥ì˜ ìµœì†Œ ê¸¸ì´ë¥¼ ì§€ì •í•©ë‹ˆë‹¤. ì—¬ê¸°ì„œëŠ” 3ìœ¼ë¡œ ì„¤ì •í–ˆìŠµë‹ˆë‹¤

ê·¸ëŸ° ë‹¤ìŒ ë¹ˆ ë¦¬ìŠ¤íŠ¸ `X_r`ì™€ `y_r`ì„ ìƒì„±í•©ë‹ˆë‹¤. ì´ ë¦¬ìŠ¤íŠ¸ë“¤ì€ ê°ê° ì›ë¬¸(source)ê³¼ ë²ˆì—­ë¬¸(target)ì˜ ë¬¸ì¥ì„ ì €ì¥í•  ê²ƒì…ë‹ˆë‹¤

`corpus`ì˜ ê° ë¬¸ì¥ì— ëŒ€í•´ ë‹¤ìŒ ì‘ì—…ì„ ìˆ˜í–‰í•©ë‹ˆë‹¤:

1. ì›ë¬¸(source)ê³¼ ë²ˆì—­ë¬¸(target)ì„ ë¶„ë¦¬í•©ë‹ˆë‹¤. ì´ ê²½ìš°ì—ëŠ” íƒ­(\t)ìœ¼ë¡œ êµ¬ë¶„ë˜ì–´ ìˆìŠµë‹ˆë‹¤
2. ì›ë¬¸ê³¼ ë²ˆì—­ë¬¸ì´ ë¹„ì–´ ìˆì§€ ì•Šì€ì§€ í™•ì¸í•©ë‹ˆë‹¤. ë¹„ì–´ ìˆë‹¤ë©´ í•´ë‹¹ ë¬¸ì¥ì„ ê±´ë„ˆëœë‹ˆë‹¤
3. ì›ë¬¸ê³¼ ë²ˆì—­ë¬¸ì„ ì •ê·œí™”(normalize)í•˜ê³  ê³µë°±ìœ¼ë¡œ ë¶„ë¦¬í•©ë‹ˆë‹¤
4. ì›ë¬¸ê³¼ ë²ˆì—­ë¬¸ì˜ ê¸¸ì´ê°€ ì„¤ì •í•œ ìµœì†Œ ë° ìµœëŒ€ ê¸¸ì´ ì‚¬ì´ì— ìˆëŠ”ì§€ í™•ì¸í•©ë‹ˆë‹¤ ë§Œì•½ ê·¸ë ‡ë‹¤ë©´, `X_r`ê³¼ `y_r`ì— ê°ê° ì›ë¬¸ê³¼ ë²ˆì—­ë¬¸ì„ ì¶”ê°€í•©ë‹ˆë‹¤

````python
MIN_LENGTH = 3
MAX_LENGTH = 25

X_r, y_r = [], [] # ì›ì‹œ ë°ì´í„°ë¥¼ ì €ì¥í•  ë¹ˆ ë¦¬ìŠ¤íŠ¸ ìƒì„±

# ë°ì´í„° ì „ì²˜ë¦¬
for parallel in corpus:
    so,ta,_ = parallel[:-1].split('\t') # ì…ë ¥ ë¬¸ì¥ê³¼ ì¶œë ¥ ë¬¸ì¥ì„ ë¶„ë¦¬
    if so.strip() == "" or ta.strip() == "": # ë§Œì•½ ì…ë ¥ ë¬¸ì¥ì´ë‚˜ ì¶œë ¥ ë¬¸ì¥ì´ ë¹„ì–´ìˆìœ¼ë©´ ë¬´ì‹œ
        continue

    normalized_so = normalize_string(so).split() # ì…ë ¥ ë¬¸ì¥ì„ ì •ê·œí™”í•˜ê³  ë‹¨ì–´ ë‹¨ìœ„ë¡œ ë¶„ë¦¬
    normalized_ta = normalize_string(ta).split() # ì¶œë ¥ ë¬¸ì¥ì„ ì •ê·œí™”í•˜ê³  ë‹¨ì–´ ë‹¨ìœ„ë¡œ ë¶„ë¦¬

    # ì…ë ¥ ë¬¸ì¥ê³¼ ì¶œë ¥ ë¬¸ì¥ì˜ ê¸¸ì´ê°€ ìµœì†Œ ê¸¸ì´ì™€ ìµœëŒ€ ê¸¸ì´ ì‚¬ì´ì— ìˆìœ¼ë©´ ë¦¬ìŠ¤íŠ¸ì— ì¶”ê°€
    if len(normalized_so) >= MIN_LENGTH and len(normalized_so) <= MAX_LENGTH \
    and len(normalized_ta) >= MIN_LENGTH and len(normalized_ta) <= MAX_LENGTH:
        X_r.append(normalized_so)
        y_r.append(normalized_ta)

print(len(X_r), len(y_r)) # ì…ë ¥ ë¬¸ì¥ê³¼ ì¶œë ¥ ë¬¸ì¥ì˜ ê°œìˆ˜ ì¶œë ¥
print(X_r[0], y_r[0]) # ì²« ë²ˆì§¸ ì…ë ¥ ë¬¸ì¥ê³¼ ì¶œë ¥ ë¬¸ì¥ ì¶œë ¥
````

##### ì†ŒìŠ¤ ë° íƒ€ê²Ÿ ë‹¨ì–´ì¥ (Vocabulary) ìƒì„±

##### ì¸ë±ìŠ¤ ë§¤í•‘ ìƒì„±
ë‹¨ì–´ì¥ì„ ê¸°ë°˜ìœ¼ë¡œ ë‹¨ì–´ì™€ ì¸ë±ìŠ¤ ì‚¬ì´ì˜ ë§¤í•‘ì„ ìƒì„±í•©ë‹ˆë‹¤. ì´ë¥¼ ìœ„í•´ ë¨¼ì € `<PAD>`, `<UNK>`, `<s>`, `</s>`ì™€ ê°™ì€ íŠ¹ìˆ˜ í† í°ì„ í¬í•¨í•œ ì´ˆê¸° ë§¤í•‘ ë”•ì…”ë„ˆë¦¬ë¥¼ ìƒì„±í•©ë‹ˆë‹¤. ê°ê°ì˜ íŠ¹ìˆ˜ í† í°ì€ ë‹¤ìŒê³¼ ê°™ì€ ì—­í• ì„ í•©ë‹ˆë‹¤:
- `<PAD>` : íŒ¨ë”©(Padding) í† í°ìœ¼ë¡œ, ë°°ì¹˜ ì²˜ë¦¬ë¥¼ ìœ„í•´ ì„œë¡œ ë‹¤ë¥¸ ê¸¸ì´ì˜ ì‹œí€€ìŠ¤ë¥¼ ë™ì¼í•œ ê¸¸ì´ë¡œ ë§ì¶”ê¸° ìœ„í•´ ì‚¬ìš©
- `<UNK>` : ì•Œ ìˆ˜ ì—†ëŠ”(Unknown) í† í°ìœ¼ë¡œ, ë‹¨ì–´ì¥ì— ì—†ëŠ” ë‹¨ì–´ë¥¼ ëŒ€ì²´í•˜ëŠ” ë° ì‚¬ìš©
- `<s>` : ì‹œì‘(Start) í† í°ìœ¼ë¡œ, ë””ì½”ë”ì˜ ì…ë ¥ ì‹œí€€ìŠ¤ì˜ ì‹œì‘ì„ ë‚˜íƒ€ëƒ„
- `</s>` : ì¢…ë£Œ(End) í† í°ìœ¼ë¡œ, ë””ì½”ë”ì˜ ì¶œë ¥ ì‹œí€€ìŠ¤ì˜ ì¢…ë£Œë¥¼ ë‚˜íƒ€ëƒ„

`source2index`ì™€ `target2index` ë”•ì…”ë„ˆë¦¬ì— ê°ê° ì†ŒìŠ¤ì™€ íƒ€ê²Ÿ ë‹¨ì–´ì¥ì˜ ë‹¨ì–´ë“¤ì„ ì¶”ê°€í•©ë‹ˆë‹¤. ì´ ë•Œ, ì´ë¯¸ ë”•ì…”ë„ˆë¦¬ì— ì¡´ì¬í•˜ëŠ” ë‹¨ì–´ëŠ” ì¶”ê°€í•˜ì§€ ì•ŠìŠµë‹ˆë‹¤.

ë§ˆì§€ë§‰ìœ¼ë¡œ, `index2source`ì™€ `index2target` ë”•ì…”ë„ˆë¦¬ë¥¼ ìƒì„±í•˜ì—¬ ì¸ë±ìŠ¤ì™€ ë‹¨ì–´ ì‚¬ì´ì˜ ì—­ë§¤í•‘ì„ êµ¬ì„±í•©ë‹ˆë‹¤. ì´ë ‡ê²Œ ìƒì„±ëœ ë§¤í•‘ ë”•ì…”ë„ˆë¦¬ë“¤ì€ ì¶”í›„ì— ë°ì´í„° ì „ì²˜ë¦¬ ë° ê²°ê³¼ í•´ì„ì— ì‚¬ìš©ë©ë‹ˆë‹¤

````python
# ì†ŒìŠ¤ì™€ íƒ€ê²Ÿ ì–´íœ˜ì§‘ì„ ìƒì„±
source_vocab = list(set(flatten(X_r))) # ì†ŒìŠ¤ ì–´íœ˜ì§‘ ìƒì„±
target_vocab = list(set(flatten(y_r))) # íƒ€ê²Ÿ ì–´íœ˜ì§‘ ìƒì„±
print(len(source_vocab), len(target_vocab)) # ì†ŒìŠ¤ ì–´íœ˜ì§‘ê³¼ íƒ€ê²Ÿ ì–´íœ˜ì§‘ì˜ í¬ê¸° ì¶œë ¥

# ì†ŒìŠ¤ì™€ íƒ€ê²Ÿ ì–´íœ˜ì§‘ì„ ì¸ë±ìŠ¤ë¡œ ë³€í™˜í•˜ëŠ” ë”•ì…”ë„ˆë¦¬ ìƒì„±
source2index = {'<PAD>': 0, '<UNK>': 1, '<s>': 2, '</s>': 3} # ì†ŒìŠ¤ ì–´íœ˜ì§‘ì„ ì¸ë±ìŠ¤ë¡œ ë³€í™˜í•˜ëŠ” ë”•ì…”ë„ˆë¦¬ ì´ˆê¸°í™”
for vo in source_vocab: # ì†ŒìŠ¤ ì–´íœ˜ì§‘ì˜ ê° ë‹¨ì–´ì— ëŒ€í•´
    if source2index.get(vo) is None: # í•´ë‹¹ ë‹¨ì–´ê°€ ë”•ì…”ë„ˆë¦¬ì— ì—†ë‹¤ë©´
        source2index[vo] = len(source2index) # ë”•ì…”ë„ˆë¦¬ì— ì¶”ê°€
index2source = {v:k for k, v in source2index.items()} # ì¸ë±ìŠ¤ë¥¼ ì†ŒìŠ¤ ì–´íœ˜ì§‘ìœ¼ë¡œ ë³€í™˜í•˜ëŠ” ë”•ì…”ë„ˆë¦¬ ìƒì„±

target2index = {'<PAD>': 0, '<UNK>': 1, '<s>': 2, '</s>': 3} # íƒ€ê²Ÿ ì–´íœ˜ì§‘ì„ ì¸ë±ìŠ¤ë¡œ ë³€í™˜í•˜ëŠ” ë”•ì…”ë„ˆë¦¬ ì´ˆê¸°í™”
for vo in target_vocab: # íƒ€ê²Ÿ ì–´íœ˜ì§‘ì˜ ê° ë‹¨ì–´ì— ëŒ€í•´
    if target2index.get(vo) is None: # í•´ë‹¹ ë‹¨ì–´ê°€ ë”•ì…”ë„ˆë¦¬ì— ì—†ë‹¤ë©´
        target2index[vo] = len(target2index) # ë”•ì…”ë„ˆë¦¬ì— ì¶”ê°€
index2target = {v:k for k, v in target2index.items()} # ì¸ë±ìŠ¤ë¥¼ íƒ€ê²Ÿ ì–´íœ˜ì§‘ìœ¼ë¡œ ë³€í™˜í•˜ëŠ” ë”•ì…”ë„ˆë¦¬ ìƒì„±
````

##### prepare_sequence í•¨ìˆ˜ì™€ ë°ì´í„° ì „ì²˜ë¦¬

- `prepare_sequence` í•¨ìˆ˜ëŠ” ì£¼ì–´ì§„ ì‹œí€€ìŠ¤ë¥¼ ì¸ë±ìŠ¤ë¡œ ë³€í™˜í•˜ëŠ” í•¨ìˆ˜ì…ë‹ˆë‹¤. ì´ í•¨ìˆ˜ëŠ” seq(ë‹¨ì–´ ì‹œí€€ìŠ¤)ì™€ to_index(ë‹¨ì–´ë¥¼ ì¸ë±ìŠ¤ë¡œ ë§¤í•‘í•˜ëŠ” ë”•ì…”ë„ˆë¦¬)ë¥¼ ì…ë ¥ìœ¼ë¡œ ë°›ì•„, ì‹œí€€ìŠ¤ì˜ ê° ë‹¨ì–´ë¥¼ í•´ë‹¹ ì¸ë±ìŠ¤ë¡œ ë³€í™˜í•˜ê³ , ì´ë¥¼ PyTorchì˜ Variableë¡œ ë³€í™˜í•˜ì—¬ ë°˜í™˜í•©ë‹ˆë‹¤. ë§Œì•½ ë‹¨ì–´ê°€ to_index ë”•ì…”ë„ˆë¦¬ì— ì—†ìœ¼ë©´, "<UNK>"(unknown) í† í°ì˜ ì¸ë±ìŠ¤ë¡œ ë³€í™˜í•©ë‹ˆë‹¤

- ì´í›„, ì „ì²˜ë¦¬ëœ ì›ë¬¸(X_r)ê³¼ ë²ˆì—­ë¬¸(y_r) ë°ì´í„°ë¥¼ ì‚¬ìš©í•˜ì—¬, ê° ì‹œí€€ìŠ¤ì˜ ëì— "</s>"(end of sentence) í† í°ì„ ì¶”ê°€í•˜ê³ , prepare_sequence í•¨ìˆ˜ë¥¼ ì‚¬ìš©í•˜ì—¬ ì¸ë±ìŠ¤ë¡œ ë³€í™˜í•œ í›„, ì´ë¥¼ PyTorchì˜ Variableë¡œ ë³€í™˜í•©ë‹ˆë‹¤. ì´ë ‡ê²Œ ë³€í™˜ëœ ë°ì´í„°ë¥¼ X_pì™€ y_p ë¦¬ìŠ¤íŠ¸ì— ì €ì¥í•©ë‹ˆë‹¤

- ë§ˆì§€ë§‰ìœ¼ë¡œ, ë³€í™˜ëœ ì›ë¬¸(X_p)ê³¼ ë²ˆì—­ë¬¸(y_p) ë°ì´í„°ë¥¼ zip í•¨ìˆ˜ë¥¼ ì‚¬ìš©í•˜ì—¬ ìŒìœ¼ë¡œ ë¬¶ì–´ train_data ë¦¬ìŠ¤íŠ¸ì— ì €ì¥í•©ë‹ˆë‹¤. ì´ë ‡ê²Œ ìƒì„±ëœ train_dataëŠ” ì›ë¬¸ê³¼ ë²ˆì—­ë¬¸ì˜ ìŒìœ¼ë¡œ ì´ë£¨ì–´ì§„ í•™ìŠµ ë°ì´í„°ì…ë‹ˆë‹¤

````python
# ì‹œí€€ìŠ¤ë¥¼ ì¤€ë¹„í•˜ëŠ” í•¨ìˆ˜
def prepare_sequence(seq, to_index):
    idxs = list(map(lambda w: to_index[w] if to_index.get(w) is not None else to_index["<UNK>"], seq)) # ì‹œí€€ìŠ¤ë¥¼ ì¸ë±ìŠ¤ë¡œ ë³€í™˜
    return Variable(LongTensor(idxs)) # ë³€í™˜ëœ ì¸ë±ìŠ¤ë¥¼ í…ì„œë¡œ ë³€í™˜

X_p, y_p = [], []

for so, ta in zip(X_r, y_r): # ê° ì†ŒìŠ¤ì™€ íƒ€ê²Ÿì— ëŒ€í•´
    X_p.append(prepare_sequence(so + ['</s>'], source2index).view(1, -1)) # ì†ŒìŠ¤ ì‹œí€€ìŠ¤ë¥¼ ì¸ë±ìŠ¤ë¡œ ë³€í™˜í•˜ì—¬ ë¦¬ìŠ¤íŠ¸ì— ì¶”ê°€
    y_p.append(prepare_sequence(ta + ['</s>'], target2index).view(1, -1)) # íƒ€ê²Ÿ ì‹œí€€ìŠ¤ë¥¼ ì¸ë±ìŠ¤ë¡œ ë³€í™˜í•˜ì—¬ ë¦¬ìŠ¤íŠ¸ì— ì¶”ê°€

train_data = list(zip(X_p, y_p)) # ì†ŒìŠ¤ì™€ íƒ€ê²Ÿ ì‹œí€€ìŠ¤ë¥¼ íŠœí”Œë¡œ ë¬¶ì–´ í•™ìŠµ ë°ì´í„° ìƒì„±

train_data[:5] # í•™ìŠµ ë°ì´í„°ì˜ ì²« 5ê°œë¥¼ ì¶œë ¥
````

- `input_lengths` : ì…ë ¥ ì‹œí€€ìŠ¤ì˜ ì‹¤ì œ ê¸¸ì´ë¥¼ ë‚˜íƒ€ë‚´ëŠ” ë¦¬ìŠ¤íŠ¸ì…ë‹ˆë‹¤. ì´ëŠ” íŒ¨ë”©ëœ ì‹œí€€ìŠ¤ë¥¼ ì²˜ë¦¬í•˜ê¸° ìœ„í•´ ì‚¬ìš©ë©ë‹ˆë‹¤

`forward` ë©”ì„œë“œëŠ” ì…ë ¥ ì‹œí€€ìŠ¤ë¥¼ ì„ë² ë”© ë ˆì´ì–´ë¥¼ í†µê³¼ì‹œí‚¤ê³ , íŒ¨ë”©ëœ ì‹œí€€ìŠ¤ë¥¼ ì²˜ë¦¬í•˜ê¸° ìœ„í•´ íŒ¨í‚¹ ë° ì–¸íŒ¨í‚¹ì„ ìˆ˜í–‰í•©ë‹ˆë‹¤. ê·¸ í›„, GRUë¥¼ í†µê³¼ì‹œì¼œ ì¶œë ¥ê³¼ ì€ë‹‰ ìƒíƒœë¥¼ ì–»ìŠµë‹ˆë‹¤. ë§ˆì§€ë§‰ìœ¼ë¡œ ì€ë‹‰ ìƒíƒœë¥¼ ë°˜í™˜í•©ë‹ˆë‹¤

````python
# ì¸ì½”ë” í´ë˜ìŠ¤ ì •ì˜
class Encoder(nn.Module):
    # ì´ˆê¸°í™” í•¨ìˆ˜
    def __init__(self, input_size, embedding_size,hidden_size, n_layers=1,bidirec=False):
        super(Encoder, self).__init__()

        # ì¸ì½”ë”ì˜ ì…ë ¥ í¬ê¸°, ì€ë‹‰ì¸µ í¬ê¸°, ë ˆì´ì–´ ìˆ˜ ì„¤ì •
        # input_size: ì…ë ¥ ë°ì´í„°ì˜ ì°¨ì› ìˆ˜
        # hidden_size: ì€ë‹‰ì¸µì˜ ë…¸ë“œ ìˆ˜
        # n_layers: ì€ë‹‰ì¸µì˜ ìˆ˜
        self.input_size = input_size
        self.hidden_size = hidden_size
        self.n_layers = n_layers

        # ì„ë² ë”© ë ˆì´ì–´ë¥¼ ì •ì˜
        # ì„ë² ë”© ë ˆì´ì–´ëŠ” ì…ë ¥ ë°ì´í„°ë¥¼ ì„ë² ë”© ë²¡í„°ë¡œ ë³€í™˜í•˜ëŠ” ì—­í• ì„ í•¨
        self.embedding = nn.Embedding(input_size, embedding_size)

        # ì–‘ë°©í–¥ GRUë¥¼ ì‚¬ìš©í• ì§€ ì—¬ë¶€ì— ë”°ë¼ GRU ë ˆì´ì–´
        # ì–‘ë°©í–¥ GRUë¥¼ ì‚¬ìš©í•˜ë©´, n_directionì„ 2ë¡œ ì„¤ì •í•˜ê³ , ê·¸ë ‡ì§€ ì•Šìœ¼ë©´ 1ë¡œ ì„¤ì •
        if bidirec:
            self.n_direction = 2
            self.gru = nn.GRU(embedding_size, hidden_size, n_layers, batch_first=True, bidirectional=True)
        else:
            self.n_direction = 1
            self.gru = nn.GRU(embedding_size, hidden_size, n_layers, batch_first=True)

    # ì€ë‹‰ ìƒíƒœë¥¼ ì´ˆê¸°í™”í•˜ëŠ” í•¨ìˆ˜
    # ì…ë ¥ì˜ í¬ê¸°ì— ë”°ë¼ 0ìœ¼ë¡œ ì±„ì›Œì§„ í…ì„œë¥¼ ìƒì„±í•˜ê³ , CUDAë¥¼ ì‚¬ìš©í•  ê²½ìš° GPUë¡œ ì´ë™
    def init_hidden(self, inputs):
        hidden = Variable(torch.zeros(self.n_layers * self.n_direction, inputs.size(0), self.hidden_size))
        return hidden.cuda() if USE_CUDA else hidden

    # ê°€ì¤‘ì¹˜ë¥¼ ì´ˆê¸°í™”í•˜ëŠ” í•¨ìˆ˜
    # ì„ë² ë”© ë ˆì´ì–´ì™€ GRU ë ˆì´ì–´ì˜ ê°€ì¤‘ì¹˜ë¥¼ xavier_uniform ë°©ì‹ìœ¼ë¡œ ì´ˆê¸°í™”
    def init_weight(self):
        self.embedding.weight = nn.init.xavier_uniform(self.embedding.weight)
        self.gru.weight_hh_l0 = nn.init.xavier_uniform(self.gru.weight_hh_l0)
        self.gru.weight_ih_l0 = nn.init.xavier_uniform(self.gru.weight_ih_l0)

    # ìˆœì „íŒŒ ì •ì˜
    # ì…ë ¥ê³¼ ì…ë ¥ ê¸¸ì´ë¥¼ ë°›ì•„ì„œ ì„ë² ë”© ë ˆì´ì–´ì™€ GRU ë ˆì´ì–´ë¥¼ í†µê³¼
    # GRU ë ˆì´ì–´ì˜ ì¶œë ¥ê³¼ ì€ë‹‰ ìƒíƒœë¥¼ ë°˜í™˜
    def forward(self, inputs, input_lengths):
        hidden = self.init_hidden(inputs)

        embedded = self.embedding(inputs)
        packed = torch.nn.utils.rnn.pack_padded_sequence(embedded, input_lengths, batch_first=True)
        outputs, hidden = self.gru(packed, hidden)
        outputs, output_lengths = torch.nn.utils.rnn.pad_packed_sequence(outputs, batch_first=True)

        # ë ˆì´ì–´ ìˆ˜ê°€ 1ë³´ë‹¤ í¬ë©´, ë§ˆì§€ë§‰ ë‘ ë ˆì´ì–´ì˜ ì€ë‹‰ ìƒíƒœë¥¼ ì‚¬ìš©í•˜ê³ , ê·¸ë ‡ì§€ ì•Šìœ¼ë©´ ë§ˆì§€ë§‰ ë ˆì´ì–´ì˜ ì€ë‹‰ ìƒíƒœë¥¼ ì‚¬ìš©
        if self.n_layers > 1:
            if self.n_direction == 2:
                hidden = hidden[-2:]
            else:
                hidden = hidden[-1]

        return outputs, torch.cat([h for h in hidden], 1).unsqueeze(1)
````

- `BATCH_SIZE` : í•™ìŠµ ì‹œ ì‚¬ìš©í•  ë°°ì¹˜ í¬ê¸°ë¥¼ ì§€ì •í•©ë‹ˆë‹¤. ì—¬ê¸°ì„œëŠ” 64ë¡œ ì„¤ì •í•˜ì˜€ìŠµë‹ˆë‹¤

- `EMBEDDING_SIZE` : ì„ë² ë”© ë²¡í„°ì˜ ì°¨ì› í¬ê¸°ë¥¼ ì§€ì •í•©ë‹ˆë‹¤. ì—¬ê¸°ì„œëŠ” 300ìœ¼ë¡œ ì„¤ì •í•˜ì˜€ìŠµë‹ˆë‹¤

- `HIDDEN_SIZE` : ì¸ì½”ë”ì˜ ì€ë‹‰ ìƒíƒœ í¬ê¸°ë¥¼ ì§€ì •í•©ë‹ˆë‹¤. ì—¬ê¸°ì„œëŠ” 512ë¡œ ì„¤ì •í•˜ì˜€ìŠµë‹ˆë‹¤

ì¸ì½”ë” ê°ì²´ë¥¼ ìƒì„±í•˜ê³  ê°€ì¤‘ì¹˜ë¥¼ ì´ˆê¸°í™”í•©ë‹ˆë‹¤:

- `encoder = Encoder(len(source2index), EMBEDDING_SIZE, HIDDEN_SIZE, 3, True).cuda()`: ì¸ì½”ë” ê°ì²´ë¥¼ ìƒì„±í•˜ë©°, ì…ë ¥ í¬ê¸°, ì„ë² ë”© í¬ê¸°, ì€ë‹‰ ìƒíƒœ í¬ê¸°, ì¸µì˜ ê°œìˆ˜, ì–‘ë°©í–¥ ì—¬ë¶€ë¥¼ ì¸ìë¡œ ì „ë‹¬í•©ë‹ˆë‹¤. `.cuda()`ë¥¼ ì‚¬ìš©í•˜ì—¬ GPUì—ì„œ ì‹¤í–‰ë˜ë„ë¡ ì„¤ì •í•©ë‹ˆë‹¤

- `encoder.init_weight()`: ì¸ì½”ë”ì˜ ê°€ì¤‘ì¹˜ë¥¼ ì´ˆê¸°í™”í•©ë‹ˆë‹¤

í•™ìŠµ ë°ì´í„°ì—ì„œ ë°°ì¹˜ë¥¼ ê°€ì ¸ì™€ ì…ë ¥ ë° íƒ€ê²Ÿ ì‹œí€€ìŠ¤ë¥¼ ì²˜ë¦¬í•©ë‹ˆë‹¤:

- `ex = next(iter(getBatch(BATCH_SIZE, train_data)))`: í•™ìŠµ ë°ì´í„°ì—ì„œ ë°°ì¹˜ë¥¼ ê°€ì ¸ì˜µë‹ˆë‹¤.

- `inputs, targets, input_lengths, target_lengths=pad_to_batch(ex, source2index, target2index)`: ê°€ì ¸ì˜¨ ë°°ì¹˜ë¥¼ ì²˜ë¦¬í•˜ì—¬ ì…ë ¥ ì‹œí€€ìŠ¤, íƒ€ê²Ÿ ì‹œí€€ìŠ¤, ì…ë ¥ ì‹œí€€ìŠ¤ ê¸¸ì´, íƒ€ê²Ÿ ì‹œí€€ìŠ¤ ê¸¸ì´ë¥¼ ì–»ìŠµë‹ˆë‹¤

ì¸ì½”ë”ë¥¼ ì‚¬ìš©í•˜ì—¬ ì…ë ¥ ì‹œí€€ìŠ¤ë¥¼ ì²˜ë¦¬í•˜ê³  ì¶œë ¥ ë° ì€ë‹‰ ìƒíƒœë¥¼ ì–»ìŠµë‹ˆë‹¤:

- `output, hidden_c = encoder(inputs,input_lengths)`: ì¸ì½”ë”ë¥¼ ì‚¬ìš©í•˜ì—¬ ì…ë ¥ ì‹œí€€ìŠ¤ë¥¼ ì²˜ë¦¬í•˜ê³  ì¶œë ¥ ë° ì€ë‹‰ ìƒíƒœë¥¼ ì–»ìŠµë‹ˆë‹¤

````python
BATCH_SIZE = 64 # BATCH_SIZEëŠ” í•œ ë²ˆì— í•™ìŠµí•  ë°ì´í„°ì˜ ìˆ˜ë¥¼ ì˜ë¯¸
EMBEDDING_SIZE = 300 # EMBEDDING_SIZEëŠ” ì„ë² ë”© ë²¡í„°ì˜ ì°¨ì› ìˆ˜ë¥¼ ì˜ë¯¸
HIDDEN_SIZE = 512 # HIDDEN_SIZEëŠ” íˆë“  ë ˆì´ì–´ì˜ ë…¸ë“œ ìˆ˜ë¥¼ ì˜ë¯¸

# ì¸ì½”ë” ê°ì²´ ìƒì„±
encoder = Encoder(
    input_size=len(source2index),
    embedding_size=EMBEDDING_SIZE,
    hidden_size=HIDDEN_SIZE,
    n_layers=3,
    bidirec=True
).cuda()

# ì¸ì½”ë”ì˜ ê°€ì¤‘ì¹˜ ì´ˆê¸°í™”
encoder.init_weight()

# ë°°ì¹˜ ë°ì´í„° ì¶”ì¶œ
ex = next(iter(getBatch(BATCH_SIZE, train_data)))
inputs, targets, input_lengths, target_lengths = pad_to_batch(ex, source2index, target2index)

# ì¸ì½”ë”ë¥¼ í†µí•´ ì¶œë ¥ê³¼ íˆë“  ìƒíƒœ ì–»ìŒ
output, hidden_c = encoder(
    inputs=inputs,
    input_lengths=input_lengths
)

print(output.size()) # [Batch_size, Sequence_length, Hidden_size]
````

#### Decoder êµ¬í˜„


DecoderëŠ” Seq2Seq ëª¨ë¸ì˜ ì¶œë ¥ ì‹œí€€ìŠ¤ë¥¼ ìƒì„±í•˜ëŠ” ë¶€ë¶„ì…ë‹ˆë‹¤. ì£¼ì–´ì§„ ì¸ì½”ë”ì˜ ì¶œë ¥ê³¼ ì´ì „ ì‹œì ì˜ ì¶œë ¥ì„ ê¸°ë°˜ìœ¼ë¡œ ë‹¤ìŒ ì‹œì ì˜ ì¶œë ¥ì„ ìƒì„±í•©ë‹ˆë‹¤. ì´ ê³¼ì •ì—ì„œ ì–´í…ì…˜ ë©”ì»¤ë‹ˆì¦˜ì„ ì‚¬ìš©í•˜ì—¬ ì…ë ¥ ì‹œí€€ìŠ¤ì™€ ì¶œë ¥ ì‹œí€€ìŠ¤ ì‚¬ì´ì˜ ì—°ê´€ì„±ì„ í•™ìŠµí•©ë‹ˆë‹¤

ì•„ë˜ì—ì„œ êµ¬í˜„ í•  Decoder í´ë˜ìŠ¤ëŠ” ë‹¤ìŒê³¼ ê°™ì€ êµ¬ì„± ìš”ì†Œë¥¼ ê°€ì§‘ë‹ˆë‹¤:

- `__init__` : ì´ˆê¸°í™” í•¨ìˆ˜ë¡œ, ì…ë ¥ í¬ê¸°, ì„ë² ë”© í¬ê¸°, ì€ë‹‰ í¬ê¸°, ì¸µì˜ ê°œìˆ˜, ë“œë¡­ì•„ì›ƒ í™•ë¥  ë“±ì„ ì¸ìë¡œ ë°›ìŠµë‹ˆë‹¤
- `init_hidden` : ì´ˆê¸° ì€ë‹‰ ìƒíƒœë¥¼ ìƒì„±í•˜ëŠ” í•¨ìˆ˜ë¡œ, ì…ë ¥ í…ì„œë¥¼ ì¸ìë¡œ ë°›ì•„ ì€ë‹‰ ìƒíƒœë¥¼ ì´ˆê¸°í™”í•©ë‹ˆë‹¤
- `init_weight` : ê°€ì¤‘ì¹˜ë¥¼ ì´ˆê¸°í™”í•˜ëŠ” í•¨ìˆ˜ë¡œ, Xavier ì´ˆê¸°í™” ë°©ë²•ì„ ì‚¬ìš©í•˜ì—¬ ì„ë² ë”©, GRU, ì„ í˜•, ì–´í…ì…˜ ê³„ì¸µì˜ ê°€ì¤‘ì¹˜ë¥¼ ì´ˆê¸°í™”í•©ë‹ˆë‹¤
- `Attention` : ì–´í…ì…˜ ë©”ì»¤ë‹ˆì¦˜ì„ êµ¬í˜„í•˜ëŠ” í•¨ìˆ˜ë¡œ, í˜„ì¬ ì€ë‹‰ ìƒíƒœ, ì¸ì½”ë” ì¶œë ¥, ì¸ì½”ë” ë§ˆìŠ¤í‚¹ì„ ì¸ìë¡œ ë°›ì•„ ì»¨í…ìŠ¤íŠ¸ ë²¡í„°ì™€ ì–´í…ì…˜ ê°€ì¤‘ì¹˜ë¥¼ ë°˜í™˜í•©ë‹ˆë‹¤
- `forward` : ë””ì½”ë”ì˜ ìˆœì „íŒŒë¥¼ êµ¬í˜„í•˜ëŠ” í•¨ìˆ˜ë¡œ, ì…ë ¥ í…ì„œ, ì»¨í…ìŠ¤íŠ¸ ë²¡í„°, ìµœëŒ€ ê¸¸ì´, ì¸ì½”ë” ì¶œë ¥, ì¸ì½”ë” ë§ˆìŠ¤í‚¹, í•™ìŠµ ì—¬ë¶€ë¥¼ ì¸ìë¡œ ë°›ì•„ ì˜ˆì¸¡ëœ ì¶œë ¥ì„ ë°˜í™˜í•©ë‹ˆë‹¤
- `decode` : ì£¼ì–´ì§„ ì»¨í…ìŠ¤íŠ¸ ë²¡í„°ì™€ ì¸ì½”ë” ì¶œë ¥ì„ ì‚¬ìš©í•˜ì—¬ ì¶œë ¥ ì‹œí€€ìŠ¤ë¥¼ ìƒì„±í•˜ëŠ” í•¨ìˆ˜ì…ë‹ˆë‹¤. ì´ í•¨ìˆ˜ëŠ” ë””ì½”ë”ì˜ ìˆœì „íŒŒë¥¼ ë°˜ë³µì ìœ¼ë¡œ ìˆ˜í–‰í•˜ì—¬ ì¶œë ¥ ì‹œí€€ìŠ¤ë¥¼ ìƒì„±í•˜ê³ , ì–´í…ì…˜ ê°€ì¤‘ì¹˜ë¥¼ ë°˜í™˜í•©ë‹ˆë‹¤

<br>

`Attention` í•¨ìˆ˜ëŠ” ì£¼ì–´ì§„ í˜„ì¬ ì€ë‹‰ ìƒíƒœì™€ ì¸ì½”ë” ì¶œë ¥ì„ ì‚¬ìš©í•˜ì—¬ ì–´í…ì…˜ ë©”ì»¤ë‹ˆì¦˜ì„ êµ¬í˜„í•©ë‹ˆë‹¤. ì–´í…ì…˜ ë©”ì»¤ë‹ˆì¦˜ì€ ì…ë ¥ ì‹œí€€ìŠ¤ì™€ ì¶œë ¥ ì‹œí€€ìŠ¤ ì‚¬ì´ì˜ ì—°ê´€ì„±ì„ í•™ìŠµí•˜ì—¬, ê° ì‹œì ì—ì„œ ì–´ë–¤ ì…ë ¥ ë‹¨ì–´ì— ì§‘ì¤‘í•´ì•¼ í•˜ëŠ”ì§€ë¥¼ ê²°ì •í•©ë‹ˆë‹¤

ìš°ë¦¬ëŠ” Bahdanau Attentionì„ ì‚¬ìš©í•˜ì—¬ Attention functionì„ êµ¬í˜„í•˜ê² ìŠµë‹ˆë‹¤

ì–´í…ì…˜ ë©”ì»¤ë‹ˆì¦˜ì˜ ìˆ˜ì‹ì€ ë‹¤ìŒê³¼ ê°™ìŠµë‹ˆë‹¤:

1. ì—ë„ˆì§€ ê³„ì‚°: $e_{ij} = a(s_{i-1}, h_j)$
   - $s_{i-1}$: ì´ì „ ì‹œì ì˜ ë””ì½”ë” ì€ë‹‰ ìƒíƒœ
   - $h_j$: ì¸ì½”ë” ì¶œë ¥ì˜ jë²ˆì§¸ ë‹¨ì–´ì— ëŒ€í•œ ì€ë‹‰ ìƒíƒœ
   - $a$: ì–´í…ì…˜ í•¨ìˆ˜ (ì—¬ê¸°ì„œëŠ” ì„ í˜• ê³„ì¸µ `self.attn`ì„ ì‚¬ìš©)

2. ì–´í…ì…˜ ê°€ì¤‘ì¹˜ ê³„ì‚°: $\alpha_{ij} = \frac{\exp(e_{ij})}{\sum_{k=1}^T \exp(e_{ik})}$
   - $\alpha_{ij}$: ië²ˆì§¸ ë””ì½”ë” ì¶œë ¥ê³¼ jë²ˆì§¸ ì¸ì½”ë” ì…ë ¥ ì‚¬ì´ì˜ ì–´í…ì…˜ ê°€ì¤‘ì¹˜

3. ì»¨í…ìŠ¤íŠ¸ ë²¡í„° ê³„ì‚°: $c_i = \sum_{j=1}^T \alpha_{ij} h_j$
   - $c_i$: ië²ˆì§¸ ë””ì½”ë” ì¶œë ¥ì— ëŒ€í•œ ì»¨í…ìŠ¤íŠ¸ ë²¡í„°

````python
class Decoder(nn.Module):
    # ì´ˆê¸°í™” í•¨ìˆ˜
    def __init__(self, input_size, embedding_size, hidden_size, n_layers=1, dropout_p=0.1):
        super(Decoder, self).__init__()

        # ë””ì½”ë”ì˜ ì€ë‹‰ì¸µ í¬ê¸°ì™€ ë ˆì´ì–´ ìˆ˜ë¥¼ ì„¤ì •
        self.hidden_size = hidden_size
        self.n_layers = n_layers

        # ì„ë² ë”© ë ˆì´ì–´ë¥¼ ìƒì„±. ì…ë ¥ í¬ê¸°ì™€ ì„ë² ë”© í¬ê¸°ë¥¼ ì¸ìë¡œ ë°›ìŒ
        self.embedding = nn.Embedding(input_size, embedding_size)
        # ë“œë¡­ì•„ì›ƒ ë ˆì´ì–´ë¥¼ ìƒì„±. ë“œë¡­ì•„ì›ƒ í™•ë¥ ì„ ì¸ìë¡œ ë°›ìŒ
        self.dropout = nn.Dropout(dropout_p)
        # GRU ë ˆì´ì–´ë¥¼ ìƒì„±. ì„ë² ë”© í¬ê¸° + ì€ë‹‰ì¸µ í¬ê¸°, ì€ë‹‰ì¸µ í¬ê¸°, ë ˆì´ì–´ ìˆ˜ë¥¼ ì¸ìë¡œ ë°›ìŒ
        self.gru = nn.GRU(embedding_size + hidden_size, hidden_size, n_layers, batch_first=True)
        # ì„ í˜• ë ˆì´ì–´ë¥¼ ìƒì„±. ì€ë‹‰ì¸µ í¬ê¸° * 2, ì…ë ¥ í¬ê¸°ë¥¼ ì¸ìë¡œ ë°›ìŒ
        self.linear = nn.Linear(hidden_size * 2, input_size)
        # Attention ë ˆì´ì–´ë¥¼ ìƒì„±. ì€ë‹‰ì¸µ í¬ê¸°ë¥¼ ì¸ìë¡œ ë°›ìŒ
        self.attn = nn.Linear(self.hidden_size, self.hidden_size)

    # ì€ë‹‰ ìƒíƒœë¥¼ ì´ˆê¸°í™”í•˜ëŠ” í•¨ìˆ˜
    def init_hidden(self,inputs):
        # ì€ë‹‰ ìƒíƒœë¥¼ 0ìœ¼ë¡œ ì´ˆê¸°í™”
        hidden = Variable(torch.zeros(self.n_layers, inputs.size(0), self.hidden_size))
        # CUDAë¥¼ ì‚¬ìš©í•  ê²½ìš° GPUë¡œ ì´ë™
        return hidden.cuda() if USE_CUDA else hidden

    # ê°€ì¤‘ì¹˜ë¥¼ ì´ˆê¸°í™”í•˜ëŠ” í•¨ìˆ˜
    def init_weight(self):
        # ê° ë ˆì´ì–´ì˜ ê°€ì¤‘ì¹˜ë¥¼ Xavier ì´ˆê¸°í™”ë¥¼ ì‚¬ìš©í•˜ì—¬ ì´ˆê¸°í™”
        self.embedding.weight = nn.init.xavier_uniform(self.embedding.weight)
        self.gru.weight_hh_l0 = nn.init.xavier_uniform(self.gru.weight_hh_l0)
        self.gru.weight_ih_l0 = nn.init.xavier_uniform(self.gru.weight_ih_l0)
        self.linear.weight = nn.init.xavier_uniform(self.linear.weight)
        self.attn.weight = nn.init.xavier_uniform(self.attn.weight)

    # Attention ë©”ì»¤ë‹ˆì¦˜ì„ êµ¬í˜„í•˜ëŠ” í•¨ìˆ˜
    def Attention(self, hidden, encoder_outputs, encoder_maskings):
        """
        hidden : 1,B,D
        encoder_outputs : B,T,D
        encoder_maskings : B,T # ByteTensor
        """
        # hiddenì˜ ì°¨ì›ì„ ë³€ê²½
        hidden = hidden[0].unsqueeze(2)  # (1,B,D) -> (B,D,1)

        # encoder_outputsì˜ í¬ê¸°ë¥¼ ê°€ì ¸ì˜´
        batch_size = encoder_outputs.size(0) # B
        max_len = encoder_outputs.size(1) # T
        # attention ì—ë„ˆì§€ë¥¼ ê³„ì‚°
        energies = self.attn(encoder_outputs.contiguous().view(batch_size * max_len, -1)) # B*T,D -> B*T,D
        energies = energies.view(batch_size,max_len, -1) # B,T,D
        attn_energies = energies.bmm(hidden).squeeze(2) # B,T,D * B,D,1 --> B,T

        # softmaxë¥¼ ì‚¬ìš©í•˜ì—¬ attention ê°€ì¤‘ì¹˜ë¥¼ ê³„ì‚°
        alpha = F.softmax(attn_energies,1) # B,T
        alpha = alpha.unsqueeze(1) # B,1,T
        # context ë²¡í„°ë¥¼ ê³„ì‚°
        context = alpha.bmm(encoder_outputs) # B,1,T * B,T,D => B,1,D

        return context, alpha

# ìˆœì „íŒŒ í•¨ìˆ˜
    def forward(self, inputs, context, max_length, encoder_outputs, encoder_maskings=None, is_training=False):
        """
        inputs : B,1 (LongTensor, ì‹œì‘ ì‹¬ë³¼)
        context : B,1,D (FloatTensor, ë§ˆì§€ë§‰ ì¸ì½”ë” ì€ë‹‰ ìƒíƒœ)
        max_length : int, ë””ì½”ë”©í•  ìµœëŒ€ ê¸¸ì´
        encoder_outputs : B,T,D
        encoder_maskings : B,T # ByteTensor
        is_training : bool, ë“œë¡­ì•„ì›ƒì„ í›ˆë ¨ ë‹¨ê³„ì—ì„œë§Œ ì ìš©í•˜ê¸° ìœ„í•¨
        """
        # ì…ë ¥ ë‹¨ì–´ì˜ ì„ë² ë”©ì„ ê³„ì‚°
        embedded = self.embedding(inputs)
        # ì€ë‹‰ ìƒíƒœë¥¼ ì´ˆê¸°í™”
        hidden = self.init_hidden(inputs)
        # í›ˆë ¨ ë‹¨ê³„ì—ì„œë§Œ ë“œë¡­ì•„ì›ƒ ì ìš©
        if is_training:
            embedded = self.dropout(embedded)

        decode = []
        # GRUë¥¼ ì ìš©
        for i in range(max_length):
            # GRUì˜ ì…ë ¥ìœ¼ë¡œ embeddedì™€ contextë¥¼ ì—°ê²°í•œ ê²ƒì„ ì‚¬ìš©
            _, hidden = self.gru(torch.cat((embedded, context), 2), hidden) # h_t = f(h_{t-1},y_{t-1},c)
            # hiddenê³¼ contextë¥¼ ì—°ê²°í•˜ì—¬ concatedë¥¼ ìƒì„±
            concated = torch.cat((hidden, context.transpose(0, 1)), 2) # y_t = g(h_t,y_{t-1},c)
            # ì„ í˜• ë ˆì´ì–´ë¥¼ í†µí•´ scoreë¥¼ ê³„ì‚°
            score = self.linear(concated.squeeze(0))
            # scoreì— softmaxë¥¼ ì ìš©í•˜ì—¬ í™•ë¥  ë¶„í¬ë¥¼ ì–»ìŒ
            softmaxed = F.log_softmax(score,1)
            # softmaxedë¥¼ decode ë¦¬ìŠ¤íŠ¸ì— ì¶”ê°€
            decode.append(softmaxed)
            # softmaxedì˜ ìµœëŒ€ê°’ ì¸ë±ìŠ¤ë¥¼ decodedì— ì €ì¥
            decoded = softmaxed.max(1)[1]
            # decodedë¥¼ ì„ë² ë”©í•˜ì—¬ embeddedë¥¼ ì—…ë°ì´íŠ¸
            embedded = self.embedding(decoded).unsqueeze(1) # y_{t-1}
            # í›ˆë ¨ ë‹¨ê³„ì—ì„œë§Œ ë“œë¡­ì•„ì›ƒ ì ìš©
            if is_training:
                embedded = self.dropout(embedded)

            # attentionì„ ì‚¬ìš©í•˜ì—¬ ë‹¤ìŒ context ë²¡í„°ë¥¼ ê³„ì‚°
            context, alpha = self.Attention(hidden, encoder_outputs, encoder_maskings)

        # decode ë¦¬ìŠ¤íŠ¸ë¥¼ í…ì„œë¡œ ë³€í™˜
        scores = torch.cat(decode, 1)
        # scoresì˜ í¬ê¸°ë¥¼ ë³€ê²½
        return scores.view(inputs.size(0) * max_length, -1)

    # ë””ì½”ë”© í•¨ìˆ˜
    def decode(self, context, encoder_outputs):
        # ë””ì½”ë”©ì„ ì‹œì‘í•˜ëŠ” ì‹¬ë³¼ì„ start_decodeì— ì €ì¥
        start_decode = Variable(LongTensor([[target2index['<s>']] * 1])).transpose(0, 1)
        # start_decodeë¥¼ ì„ë² ë”©
        embedded = self.embedding(start_decode)
        # ì€ë‹‰ ìƒíƒœë¥¼ ì´ˆê¸°í™”
        hidden = self.init_hidden(start_decode)

        decodes = []
        attentions = []
        decoded = embedded
        # decodedê°€ ì¢…ë£Œ ì‹¬ë³¼ì´ ë  ë•Œê¹Œì§€ ë°˜ë³µ
        while decoded.data.tolist()[0] != target2index['</s>']: # </s>ê¹Œì§€
            # GRUì˜ ì…ë ¥ìœ¼ë¡œ embeddedì™€ contextë¥¼ ì—°ê²°í•œ ê²ƒì„ ì‚¬ìš©
            _, hidden = self.gru(torch.cat((embedded, context), 2), hidden) # h_t = f(h_{t-1},y_{t-1},c)
            # hiddenê³¼ contextë¥¼ ì—°ê²°í•˜ì—¬ concatedë¥¼ ìƒì„±
            concated = torch.cat((hidden, context.transpose(0, 1)), 2) # y_t = g(h_t,y_{t-1},c)
            # ì„ í˜• ë ˆì´ì–´ë¥¼ í†µí•´ scoreë¥¼ ê³„ì‚°
            score = self.linear(concated.squeeze(0))
            # scoreì— softmaxë¥¼ ì ìš©í•˜ì—¬ í™•ë¥  ë¶„í¬ë¥¼ ì–»ìŒ
            softmaxed = F.log_softmax(score,1)
            # softmaxedë¥¼ decodes ë¦¬ìŠ¤íŠ¸ì— ì¶”ê°€
            decodes.append(softmaxed)
            # softmaxedì˜ ìµœëŒ€ê°’ ì¸ë±ìŠ¤ë¥¼ decodedì— ì €ì¥
            decoded = softmaxed.max(1)[1]
            # decodedë¥¼ ì„ë² ë”©í•˜ì—¬ embeddedë¥¼ ì—…ë°ì´íŠ¸
            embedded = self.embedding(decoded).unsqueeze(1) # y_{t-1}
            # attentionì„ ì‚¬ìš©í•˜ì—¬ ë‹¤ìŒ context ë²¡í„°ë¥¼ ê³„ì‚°
            context, alpha = self.Attention(hidden, encoder_outputs,None)
            # alphaë¥¼ attentions ë¦¬ìŠ¤íŠ¸ì— ì¶”ê°€
            attentions.append(alpha.squeeze(1))

        # decodes ë¦¬ìŠ¤íŠ¸ë¥¼ í…ì„œë¡œ ë³€í™˜í•˜ê³  ìµœëŒ€ê°’ ì¸ë±ìŠ¤ë¥¼ ë°˜í™˜
        return torch.cat(decodes).max(1)[1], torch.cat(attentions)
````

## 2. Seq2seq ë²ˆì—­ í•™ìŠµ

```
ğŸ’¡ ëª©ì°¨ ê°œìš” : êµ¬í˜„í•œ Seq2seqì„ í†µí•´ ë²ˆì—­ì„ í•™ìŠµí•´ë³´ê³ , ì‘ë™ ì›ë¦¬ë¥¼ ì´í•´í•©ë‹ˆë‹¤
```

Seq2Seq ëª¨ë¸ì„ í•™ìŠµí•˜ëŠ” ê³¼ì •ì„ êµ¬í˜„í•´ë´…ì‹œë‹¤. ì£¼ìš” ì„¤ì • ë³€ìˆ˜ë“¤ì€ ë‹¤ìŒê³¼ ê°™ìŠµë‹ˆë‹¤:

- `EPOCH`: ì „ì²´ ë°ì´í„°ì…‹ì— ëŒ€í•œ í•™ìŠµ ë°˜ë³µ íšŸìˆ˜
- `BATCH_SIZE`: í•œ ë²ˆì— í•™ìŠµí•  ë°ì´í„°ì˜ ê°œìˆ˜
- `EMBEDDING_SIZE`: ë‹¨ì–´ ì„ë² ë”© ë²¡í„°ì˜ ì°¨ì› í¬ê¸°
- `HIDDEN_SIZE`: ì¸ì½”ë”ì™€ ë””ì½”ë”ì˜ ì€ë‹‰ ìƒíƒœì˜ í¬ê¸°
- `LR`: í•™ìŠµë¥ (learning rate)ë¡œ, ëª¨ë¸ì˜ ê°€ì¤‘ì¹˜ë¥¼ ì—…ë°ì´íŠ¸í•˜ëŠ” ì†ë„ë¥¼ ê²°ì •í•¨
- `DECODER_LEARNING_RATIO`: ë””ì½”ë”ì˜ í•™ìŠµë¥ ì„ ì¸ì½”ë”ì˜ í•™ìŠµë¥ ê³¼ ë‹¤ë¥´ê²Œ ì„¤ì •í•  ìˆ˜ ìˆìŒ. ì—¬ê¸°ì„œëŠ” ì¸ì½”ë”ì˜ í•™ìŠµë¥ ì— 5ë¥¼ ê³±í•œ ê°’ì„ ì‚¬ìš©
- `RESCHEDULED`: í•™ìŠµë¥ ì„ ì¡°ì ˆí•˜ëŠ” ìŠ¤ì¼€ì¤„ë§ ì—¬ë¶€ë¥¼ ê²°ì • ë³€ìˆ˜

ë¨¼ì €, ì¸ì½”ë”ì™€ ë””ì½”ë” ê°ì²´ë¥¼ ìƒì„±í•˜ê³  ê°€ì¤‘ì¹˜ë¥¼ ì´ˆê¸°í™”í•©ë‹ˆë‹¤. ê·¸ëŸ° ë‹¤ìŒ, ì†ì‹¤ í•¨ìˆ˜ (loss function)ì™€ ìµœì í™” ì•Œê³ ë¦¬ì¦˜ì„ ì„¤ì •í•©ë‹ˆë‹¤

í•™ìŠµ ê³¼ì •ì€ ë‹¤ìŒê³¼ ê°™ìŠµë‹ˆë‹¤:
1. ë°°ì¹˜ ë°ì´í„°ë¥¼ ê°€ì ¸ì˜´
2. ì¸ì½”ë”ì™€ ë””ì½”ë”ì˜ ê¸°ìš¸ê¸°ë¥¼ 0ìœ¼ë¡œ ì´ˆê¸°í™”
3. ì¸ì½”ë”ë¥¼ ì‚¬ìš©í•˜ì—¬ ì…ë ¥ ë°ì´í„°ë¥¼ ì²˜ë¦¬í•˜ê³ , ì€ë‹‰ ìƒíƒœì™€ ì¶œë ¥ì„ ì–»ìŒ
4. ë””ì½”ë”ë¥¼ ì‚¬ìš©í•˜ì—¬ ì˜ˆì¸¡ê°’ì„ ìƒì„±
5. ì†ì‹¤ í•¨ìˆ˜ë¥¼ ì‚¬ìš©í•˜ì—¬ ì˜ˆì¸¡ê°’ê³¼ ì‹¤ì œ ëª©í‘œê°’ ì‚¬ì´ì˜ ì°¨ì´ë¥¼ ê³„ì‚°
6. ì†ì‹¤ì„ ì—­ì „íŒŒí•˜ì—¬ ëª¨ë¸ì˜ ê°€ì¤‘ì¹˜ë¥¼ ì—…ë°ì´íŠ¸

í•™ìŠµ ì¤‘ê°„ì— í•™ìŠµë¥ ì„ ì¡°ì ˆí•  ìˆ˜ ìˆìŠµë‹ˆë‹¤. ì—¬ê¸°ì„œëŠ” ì—í¬í¬ì˜ ì ˆë°˜ ì§€ì ì—ì„œ í•™ìŠµë¥ ì„ 0.01ë°°ë¡œ ë‚®ì¶”ëŠ” ë°©ì‹ì„ ì‚¬ìš©í•©ë‹ˆë‹¤. ì´ë¥¼ ìœ„í•´ `RESCHEDULED` ë³€ìˆ˜ë¥¼ ì‚¬ìš©í•˜ì—¬ í•™ìŠµë¥ ì´ ì¡°ì ˆë˜ì—ˆëŠ”ì§€ í™•ì¸í•˜ê³ , ì¡°ì ˆë˜ì§€ ì•Šì•˜ë‹¤ë©´ ì—í¬í¬ì˜ ì ˆë°˜ ì§€ì ì—ì„œ í•™ìŠµë¥ ì„ ë³€ê²½í•©ë‹ˆë‹¤

````python
# í•™ìŠµì— í•„ìš”í•œ íŒŒë¼ë¯¸í„° ì„¤ì •
EPOCH = 5 # ì´ í•™ìŠµí•  epochì˜ ìˆ˜
BATCH_SIZE = 256 # í•œ ë²ˆì— í•™ìŠµí•  ë°ì´í„°ì˜ ìˆ˜
EMBEDDING_SIZE = 300 # ì„ë² ë”© ë²¡í„°ì˜ ì°¨ì› ìˆ˜
HIDDEN_SIZE = 512 # íˆë“  ë ˆì´ì–´ì˜ ë…¸ë“œ ìˆ˜
LR = 0.001 # í•™ìŠµë¥ 
DECODER_LEARNING_RATIO = 5.0 # ë””ì½”ë”ì˜ í•™ìŠµë¥  ë¹„ìœ¨
RESCHEDULED = False # í•™ìŠµë¥  ì¬ì¡°ì • ì—¬ë¶€

# ì¸ì½”ë”ì™€ ë””ì½”ë” ê°ì²´ ìƒì„±
encoder = Encoder(len(source2index), EMBEDDING_SIZE, HIDDEN_SIZE, 3, True)
decoder = Decoder(len(target2index), EMBEDDING_SIZE, HIDDEN_SIZE * 2)

# ì¸ì½”ë”ì™€ ë””ì½”ë”ì˜ ê°€ì¤‘ì¹˜ ì´ˆê¸°í™”
encoder.init_weight()
decoder.init_weight()

# GPUê°€ ì‚¬ìš© ê°€ëŠ¥í•œ ê²½ìš°, ì¸ì½”ë”ì™€ ë””ì½”ë”ë¥¼ GPUë¡œ ì´ë™
if USE_CUDA:
    encoder = encoder.cuda()
    decoder = decoder.cuda()

# ì†ì‹¤ í•¨ìˆ˜ ì„¤ì • (CrossEntropyLoss ì‚¬ìš©, íŒ¨ë”©ëœ ë¶€ë¶„ì€ ë¬´ì‹œ)
loss_function = nn.CrossEntropyLoss(ignore_index=0)

# ìµœì í™” í•¨ìˆ˜ ì„¤ì • (Adam ì‚¬ìš©)
enc_optimizer = optim.Adam(encoder.parameters(), lr=LR)
dec_optimizer = optim.Adam(decoder.parameters(), lr=LR * DECODER_LEARNING_RATIO)

# í•™ìŠµ ì‹œì‘
for epoch in range(EPOCH):
    losses=[] # ì†ì‹¤ì„ ì €ì¥í•  ë¦¬ìŠ¤íŠ¸
    for i, batch in enumerate(getBatch(BATCH_SIZE, train_data)): # ë°°ì¹˜ ì‚¬ì´ì¦ˆë§Œí¼ ë°ì´í„°ë¥¼ ê°€ì ¸ì˜´
        inputs, targets, input_lengths, target_lengths = pad_to_batch(batch, source2index, target2index) # ë°°ì¹˜ ë°ì´í„°ë¥¼ íŒ¨ë”©

        # ì…ë ¥ ë°ì´í„°ì— ëŒ€í•œ ë§ˆìŠ¤í¬ ìƒì„±
        input_masks = torch.cat([Variable(ByteTensor(tuple(map(lambda s: s ==0, t.data)))) for t in inputs]).view(inputs.size(0), -1)
        start_decode = Variable(LongTensor([[target2index['<s>']] * targets.size(0)])).transpose(0, 1) # ë””ì½”ë”© ì‹œì‘ í† í° ì„¤ì •
        encoder.zero_grad() # ì¸ì½”ë”ì˜ ê·¸ë˜ë””ì–¸íŠ¸ ì´ˆê¸°í™”
        decoder.zero_grad() # ë””ì½”ë”ì˜ ê·¸ë˜ë””ì–¸íŠ¸ ì´ˆê¸°í™”
        output, hidden_c = encoder(inputs, input_lengths) # ì¸ì½”ë”ë¥¼ í†µí•´ ì¶œë ¥ê³¼ íˆë“  ìƒíƒœ ì–»ìŒ

        # ë””ì½”ë”ë¥¼ í†µí•´ ì˜ˆì¸¡ê°’ ìƒì„±
        preds = decoder(start_decode, hidden_c, targets.size(1), output, input_masks, True)

        loss = loss_function(preds, targets.view(-1)) # ì†ì‹¤ ê³„ì‚°
        losses.append(loss.data.tolist() ) # ì†ì‹¤ ì €ì¥
        loss.backward() # ì—­ì „íŒŒ ìˆ˜í–‰
        torch.nn.utils.clip_grad_norm(encoder.parameters(), 50.0) # ê·¸ë˜ë””ì–¸íŠ¸ í´ë¦¬í•‘
        torch.nn.utils.clip_grad_norm(decoder.parameters(), 50.0) # ê·¸ë˜ë””ì–¸íŠ¸ í´ë¦¬í•‘
        enc_optimizer.step() # ì¸ì½”ë” íŒŒë¼ë¯¸í„° ì—…ë°ì´íŠ¸
        dec_optimizer.step() # ë””ì½”ë” íŒŒë¼ë¯¸í„° ì—…ë°ì´íŠ¸

        # ì¼ì • ê°„ê²©ìœ¼ë¡œ ì†ì‹¤ ì¶œë ¥
        if i % 50 == 0 or i % len(train_data) == 0:
            print("[%02d/%d] [%03d/%d] mean_loss : %0.2f" %(epoch, EPOCH, i, len(train_data)//BATCH_SIZE, np.mean(losses)))
            losses=[]

    # í•™ìŠµ ì¤‘ê°„ì— í•™ìŠµë¥  ì¬ì¡°ì •
    if RESCHEDULED == False and epoch  == EPOCH//2:
        LR *= 0.01
        enc_optimizer = optim.Adam(encoder.parameters(), lr=LR)
        dec_optimizer = optim.Adam(decoder.parameters(), lr=LR * DECODER_LEARNING_RATIO)
        RESCHEDULED = True
````

## 3. ì–´í…ì…˜ ì‹œê°í™” í•¨ìˆ˜ ë° ê²°ê³¼ í™•ì¸

```
ğŸ’¡ ëª©ì°¨ ê°œìš” : í•™ìŠµëœ ì–´í…ì…˜ì„ ì‹œê°í™”í•˜ê³ , ê²°ê³¼ë¥¼ í™•ì¸í•©ë‹ˆë‹¤
```

`show_attention` í•¨ìˆ˜ëŠ” ì…ë ¥ ë‹¨ì–´, ì¶œë ¥ ë‹¨ì–´, ì–´í…ì…˜ ê°€ì¤‘ì¹˜ë¥¼ ì…ë ¥ìœ¼ë¡œ ë°›ì•„ ì–´í…ì…˜ ê°€ì¤‘ì¹˜ë¥¼ ì‹œê°í™”í•˜ëŠ” í•¨ìˆ˜ì…ë‹ˆë‹¤. ì´ í•¨ìˆ˜ë¥¼ ì‚¬ìš©í•˜ì—¬ ì…ë ¥ ë‹¨ì–´ì™€ ì¶œë ¥ ë‹¨ì–´ ì‚¬ì´ì˜ ì–´í…ì…˜ ê°€ì¤‘ì¹˜ë¥¼ ê´€ì°°í•  ìˆ˜ ìˆìŠµë‹ˆë‹¤

2. `test` ë³€ìˆ˜ì— ë¬´ì‘ìœ„ë¡œ ì„ íƒëœ í•™ìŠµ ë°ì´í„°ë¥¼ ì €ì¥í•˜ê³ , ì´ë¥¼ ì¸ì½”ë”ì™€ ë””ì½”ë”ì— ì „ë‹¬í•˜ì—¬ ë²ˆì—­ ê²°ê³¼ë¥¼ ìƒì„±í•©ë‹ˆë‹¤. ë²ˆì—­ ê²°ê³¼ì™€ í•¨ê»˜ ì–´í…ì…˜ ê°€ì¤‘ì¹˜ë„ ë°˜í™˜ë©ë‹ˆë‹¤
