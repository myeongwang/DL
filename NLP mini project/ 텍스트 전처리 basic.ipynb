### 1-2. Cleaning

> ì½”í¼ìŠ¤ ë‚´ì—ì„œ í† í°í™” ì‘ì—…ì— ë°©í•´ê°€ ë˜ê±°ë‚˜ ì˜ë¯¸ê°€ ì—†ëŠ” ë¶€ë¶„ì˜ í…ìŠ¤íŠ¸, ë…¸ì´ì¦ˆë¥¼ ì œê±°í•˜ëŠ” ì‘ì—…ì…ë‹ˆë‹¤. í† í°í™” ì „ì— ì •ì œë¥¼ í•˜ê¸°ë„ í•˜ì§€ë§Œ, ì´í›„ì—ë„ ì—¬ì „íˆ ë‚¨ì•„ìˆëŠ” ë…¸ì´ì¦ˆë“¤ì„ ì œê±°í•˜ê¸° ìœ„í•´ ì§€ì†ì ìœ¼ë¡œ ìˆ˜í–‰í•©ë‹ˆë‹¤. ì´ë•Œ ë…¸ì´ì¦ˆëŠ” íŠ¹ìˆ˜ ë¬¸ì ê°™ì€ ì•„ë¬´ ì˜ë¯¸ë„ ê°–ì§€ ì•ŠëŠ” ê¸€ìë“¤ì„ ì˜ë¯¸í•˜ê¸°ë„ í•˜ì§€ë§Œ, ë¶„ì„í•˜ê³ ì í•˜ëŠ” ëª©ì ì— ë§ì§€ ì•ŠëŠ” ë¶ˆí•„ìš”í•œ ë‹¨ì–´ë“¤ì„ ë§í•©ë‹ˆë‹¤


````python
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords') # nltk ë‚´ì¥ stopword ë‹¤ìš´ë¡œë“œ

stop=set(stopwords.words('english'))
print(stop)
````

nltkì—ì„œëŠ” ì‚¬ì „ì— ì •ì˜ëœ ì„¸íŠ¸ë¥¼ ê°–ê³ ìˆìœ¼ë©° ì‰½ê²Œ ë¡œë“œí•˜ì—¬ ì‚¬ìš©í•  ìˆ˜ ìˆìŠµë‹ˆë‹¤. ì´ë¥¼ í™œìš©í•´ ë¶ˆìš©ì–´ë¥¼ ì œì™¸í•œ ë‹¨ì–´ë“¤ì„ ì‰½ê²Œ ì¶”ì¶œí•©ë‹ˆë‹¤

````python
sen="I want to go to shopping and a I want to buy some of snack"
tokens=nltk.word_tokenize(sen)

clean_tokens=[]
for tok in tokens:
  if len(tok.lower())>1 and (tok.lower() not in stop): # í† í°ì˜ ê¸¸ì´ê°€ 1 ì´ˆê³¼ì´ê³ , stopwordì— ì†í•˜ì§€ ì•ŠëŠ” ê²½ìš°
    clean_tokens.append(tok) # clean_tokens ë¦¬ìŠ¤íŠ¸ì— ì¶”ê°€


print("ë¶ˆìš©ì–´ í¬í•¨: ",tokens)
print("ë¶ˆìš©ì–´ ë¯¸í¬í•¨: ",clean_tokens)
````

### 1-3. Normalization

> Normalizationì€ ì˜ë¯¸ê°€ ì¤‘ë³µë˜ê±°ë‚˜ ì˜ë¯¸ë¡ ì ìœ¼ë¡œ ìœ ì‚¬í•œ ë‹¨ì–´ë“¤ì„ í•˜ë‚˜ë¡œ í†µí•©í•˜ê±°ë‚˜, ë‹¨ì–´ì˜ ì›í˜•ì„ ì°¾ì•„ í†µì¼í•´ì£¼ëŠ” ì‘ì—…ì„ ë§í•©ë‹ˆë‹¤. í¬ê²Œ Stemmingê³¼ Lemmatizationì´ ìˆìŠµë‹ˆë‹¤


#### 1-3-1 Stemming
* stem:(ì‹ë¬¼ì˜) ì¤„ê¸°ë¼ëŠ” ëœ»ì„ ê°€ì§„ ì˜ì–´ë‹¨ì–´ë¡œ, ì–¸ì–´í•™ì—ì„œëŠ” stemì„ ì–´ê°„ì´ë¼ê³  í•©ë‹ˆë‹¤
* ì–´ê°„ì€ êµ´ì ˆí•˜ëŠ” ë‹¨ì–´ì—ì„œ ë³€í™”í•˜ì§€ ì•ŠëŠ” ë¶€ë¶„ì„ ì˜ë¯¸í•©ë‹ˆë‹¤. ì¦‰, stemmingì´ë€ ì–´ê°„ ì¶”ì¶œì„ ë§í•©ë‹ˆë‹¤

* ì‰½ê²Œ ë§í•´ í˜•íƒœê°€ ë³€í•œ ë‹¨ì–´ë¡œë¶€í„° êµ°ë”ë”ê¸°ë¥¼ ì œê±°í•˜ê³  ê·¸ ë‹¨ì–´ì˜ ì›ë˜ ëª¨ìŠµì„ ì¶”ì¶œí•˜ëŠ” ê²ƒì„ ë§í•©ë‹ˆë‹¤

* ì˜ˆë¥¼ ë“¤ì–´ goingì´ë¼ëŠ” ë‹¨ì–´ê°€ ìˆë‹¤ë©´ Stemmingì„ ì§„í–‰í• ì‹œ goê°€ ë©ë‹ˆë‹¤. Computersë¼ëŠ” ë‹¨ì–´ë¥¼ Stemmingì„ ì§„í–‰í•  ì‹œ Comput ë¥¼ ì¶”ì¶œí•˜ëŠ” ê³¼ì •ì„ Stemmingì´ë¼ê³  í•©ë‹ˆë‹¤


````python
from nltk.stem import PorterStemmer
from nltk.tokenize import word_tokenize
````

ìš°ì„  stemmingì„ í•˜ê¸° ì „ì— ë¬¸ì¥ë“¤ì„ ë‹¨ì–´ ë‹¨ìœ„ë¡œ í† í°í™” ì‹œì¼œì¤ë‹ˆë‹¤

````python
s = PorterStemmer() # stemmer ì„ ì–¸
text = "This was not the map we found in Billy Bones's chest, but an accurate copy, complete in all things--names and heights and soundings--with the single exception of the red crosses and the written notes."
words = word_tokenize(text)
print(words)
````

í† í°í™”ëœ ë‹¨ì–´ë“¤ì— ëŒ€í•´ì„œ stemmingì„ ì§„í–‰í•©ë‹ˆë‹¤. ì´ë•Œ stemmingì€ rule ê¸°ë°˜ì˜ ì•Œê³ ë¦¬ì¦˜ìœ¼ë¡œ ì´ë£¨ì–´ì§€ê¸° ë•Œë¬¸ì— ë“±ë¡ì´ ë˜ì–´ìˆì§€ ì•Šì€ ë‹¨ì–´ì˜ ê²½ìš°ì´ê±°ë‚˜ ì˜ëª»ì¸ì‹ë˜ëŠ” ê²½ìš° ë¶€ì •í™•í•  ìˆ˜ ìˆëŠ” ì—¬ì§€ê°€ ë‚¨ì•„ìˆìŠµë‹ˆë‹¤

````python
for i in words:
  print(f'{i} ==> {s.stem(i)}') # stemmingë˜ê¸° ì´ì „ë‹¨ì–´ì™€ ì´í›„ ë‹¨ì–´ë¥¼ ì¶œë ¥
````

#### 1-3-2 Lemmatization

* Lemmatizationì´ë€ ë¬¸ì¥ ì†ì—ì„œ ë‹¤ì–‘í•œ í˜•íƒœë¡œ í™œìš©ëœ(inflected) ë‹¨ì–´ì˜ í‘œì œì–´(lemma)ë¥¼ ì°¾ëŠ” ì¼ì„ ëœ»í•©ë‹ˆë‹¤
* ì—¬ê¸°ì„œ ë§í•˜ëŠ” í‘œì œì–´ë€ ì‚¬ì „ì—ì„œ ë‹¨ì–´ì˜ ëœ»ì„ ì°¾ì„ ë•Œ ì“°ëŠ” ê¸°ë³¸í˜•ì´ë¼ê³  ìƒê°í•˜ì‹œë©´ ë©ë‹ˆë‹¤
* ì¦‰ Lemmatizationì€ ë‹¨ì–´ì˜ ì›í˜•ì„ ì¶”ì¶œí•´ì£¼ëŠ” ë…€ì„ì…ë‹ˆë‹¤
* ì˜ˆë¥¼ë“¤ì–´, isë¥¼ Lemmatizationí•˜ë©´ beê°€ ë˜ê³ , ateì„ Lemmatizationí•˜ë©´ eatì´ ë©ë‹ˆë‹¤

````python
from nltk.stem import WordNetLemmatizer
nltk.download('wordnet')
````

Lemmatizationì„ ìœ„í•´ WordNetLemmatizerë¥¼ ë¡œë“œí•˜ê³ , ì—¬ëŸ¬ ë‹¨ì–´ë“¤ì„ ë‚˜ì—´í•´ë³´ê² ìŠµë‹ˆë‹¤

````python
lemmatizer = WordNetLemmatizer() # lemmatizer ì„ ì–¸
words = ['doing', 'has', 'going', 'loves', 'lives', 'flying', 'dies', 'watching', 'started', 'seen']
````

WordNetLemmatizerëŠ” ì…ë ¥ìœ¼ë¡œ ë‹¨ì–´ê°€ ë™ì‚¬ í’ˆì‚¬ë¼ëŠ” ì‚¬ì‹¤ì„ ì•Œë ¤ì¤„ ìˆ˜ ìˆìŠµë‹ˆë‹¤. ì¦‰, ë‹¨ì–´ë“¤ì´ ë¬¸ì¥ì—ì„œ ë™ì‚¬ë¡œ ì“°ì˜€ë‹¤ëŠ” ê²ƒì„ ì•Œë ¤ì¤€ë‹¤ë©´ í‘œì œì–´ ì¶”ì¶œê¸°ëŠ” í’ˆì‚¬ì˜ ì •ë³´ë¥¼ ë³´ì¡´í•˜ë©´ì„œ ì •í™•í•œ Lemmaë¥¼ ì¶œë ¥í•˜ê²Œ ë©ë‹ˆë‹¤

````python
for i in words:
  lemma = lemmatizer.lemmatize(i, pos='v')
  print(f'{i} ==> {lemma}') # lemmatizationë˜ê¸° ì´ì „ë‹¨ì–´ì™€ ì´í›„ ë‹¨ì–´ë¥¼ ì¶œë ¥
````

### (ì˜ˆì œ) HTML í¬ë¡¤ë§í›„ ë‹¨ì–´ ë¹ˆë„ìˆ˜ êµ¬í•˜ê¸°

````python
import nltk
import urllib
import re
from bs4 import BeautifulSoup
from nltk.corpus import stopwords
import matplotlib
import matplotlib.pyplot as plt
import random

#nltk.download('averaged_perceptron_tagger')
nltk.download('stopwords')
````

````python
response=urllib.request.urlopen('http://python.org/') # ì›¹ì— ì •ë³´ë¥¼ ìš”ì²­í•œ í›„, ëŒë ¤ë°›ì€ ì‘ë‹µì„ ì €ì¥í•˜ì—¬ â€˜ì‘ë‹µ ê°ì²´(HTTPResponse)â€™ë¥¼ ë°˜í™˜
html=response.read()
clean=BeautifulSoup(html,'html.parser').get_text() # html ì½”ë“œ ì •ì œ ì§„í–‰

tokens=[]
for tok in clean.split():
  tokens.append(tok) # tokens ë¦¬ìŠ¤íŠ¸ì— í† í°ë“¤ì„ ì¶”ê°€

stop=set(stopwords.words('english')) # ë¶ˆìš©ì–´ ì„ ì–¸

clean_tokens=[]
for tok in tokens:
  if len(tok.lower())>1 and (tok.lower() not in stop): # ê¸¸ì´ê°€ 1 ì´ˆê³¼ì´ë©° stop wordê°€ ì•„ë‹Œ ê²ƒë§Œ ì¶”ì¶œí•´
    clean_tokens.append(tok) # clean_tokens ë¦¬ìŠ¤íŠ¸ì— ì¶”ê°€

Freq_dist_nltk=nltk.FreqDist(clean_tokens) # FreqDist í´ë˜ìŠ¤ëŠ” ë¬¸ì„œì— ì‚¬ìš©ëœ ë‹¨ì–´(í† í°)ì˜ ì‚¬ìš©ë¹ˆë„ ì •ë³´ë¥¼ ë‹´ëŠ” í´ë˜ìŠ¤
Freq_dist_nltk.plot(30, cumulative=False) # ìƒìœ„ 30ê°œë¥¼ ì‹œê°í™”. ëˆ„ì  ê°¯ìˆ˜ í‘œì‹œëŠ” False
````
